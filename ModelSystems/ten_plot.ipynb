```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = np.load("num_op_array.npy")
dist = np.load("distortion.npy")
```

```python
num = [int(i) for i in num]
```

```python
c = -0.5
for i in dist:
    plt.plot(num[25:], i[25:], label=str(c))
    c -= 0.5
plt.xlabel("Number of Order Parameters", size=18)
plt.ylabel("Distortion Jump", size=18)
plt.xticks(size=12)
plt.yticks(size=12)
plt.legend(loc=1)
plt.show()
```
