```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_op = np.load("num_op_array.npy")
no_str = [int(i) for i in num_op]
dist = np.load("BUT_distortion.npy")
```

```python
c = 0.0
for i in dist:
    c -= 0.5
    plt.plot(no_str, i, label=str(c))
    if c < -2.0:
        break;
plt.xlabel("Number of Order Parameters", size=18)
plt.ylabel("Distortion Jump", size=18)
plt.xticks(size=12)
plt.yticks(size=12)
plt.axvline(x=8, color="black", linestyle="--")
plt.legend(loc=1)
plt.show()
```
