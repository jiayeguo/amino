```python
file = open("PAVAN")
file.readline()
bound = True
ubtime = 0.0
rbtime = 0.0
average = 0.0
count = 0
for line in file:
    split = line.split()
    if (float(split[1]) > 2) and bound == True:
        #print("UNBINDING " + split[0])
        #print("UB TIME = " + (str(float(split[0]) - ubtime)))
        average = average + (float(split[0]) - ubtime)
        count = count + 1
        bound = False
        rbtime = float(split[0])
    elif (float(split[1]) < 2) and bound == False:
        #print("BINDING " + split[0])
        #print("RB TIME = " + (str(float(split[0]) - rbtime)))
        bound = True
        ubtime = float(split[0])
file.close()
```

```python
import matplotlib.pyplot as plt
file = open("PAVAN")
file.readline()
x = []
count = 0.0
vals = []
for line in file:
    x.append(count)
    count = count + 0.01
    vals.append(float(line.split()[1]))
plt.plot(x, vals)
plt.xlabel("Time (nanoseconds)")
plt.ylabel("Distance between ligand and protein")
plt.show()
```
